```python
# Merging two data frames
# Creating data frames
df1 = pd.DataFrame({'key': ['k0', 'k1', 'k2', 'k3'],
                        'B': ['B0', 'B1', 'B2', 'B3'],
                        'C': ['C0', 'C1', 'C2', 'C3'],
                        'D': ['D0', 'D1', 'D2', 'D3']},
                        )
df2 = pd.DataFrame({'key': ['k8', 'k5', 'k2', 'k3'],
                        'E': ['B4', 'B5', 'B6', 'B7'],
                        'C': ['C4', 'C1', 'C6', 'C3'],
                        'F': ['D4', 'D5', 'D6', 'D7']},
                         )
df3 = pd.DataFrame({'key3': ['k1', 'k4', 'k2', 'k3'],
                        'G': ['B8', 'B9', 'B10', 'B11'],
                        'H': ['C8', 'C9', 'C10', 'C11'],
                        'I': ['D8', 'D9', 'D10', 'D11']},
                        )

df1
```

```python
df2
```

```python
df3
```

```python
mdf1 = pd.merge(df1,df2, how='left', on='key')
mdf1
```

```python
mdf2 = pd.merge(mdf1,df3, how='inner', left_on='key', right_on='key3')
mdf2
```

```python
mdf3 = pd.merge(df1,df2, on=['key','C'])
mdf3
```

```python
# Joining two data frames
# Creating data frames
df1 = pd.DataFrame({
                        'B': ['B0', 'B1', 'B2', 'B3'],
                        'C': ['C0', 'C1', 'C2', 'C3'],
                        'D': ['D0', 'D1', 'D2', 'D3']},
                        index=['k0', 'k1', 'k2', 'k3'])
df2 = pd.DataFrame({
                        'E': ['B4', 'B5', 'B6', 'B7'],
                        'C': ['C4', 'C1', 'C6', 'C3'],
                        'F': ['D4', 'D5', 'D6', 'D7']},
                         index=['k8', 'k5', 'k2', 'k3'])
df3 = pd.DataFrame({
                        'G': ['B8', 'B9', 'B10', 'B11'],
                        'H': ['C8', 'C9', 'C10', 'C11'],
                        'I': ['D8', 'D9', 'D10', 'D11']},
                        index=['k1', 'k4', 'k2', 'k3'])

df1
```

```python
df2.join(df3)
```

If coumn names overlap, a suffix should be added for unique column name

```python
df1.join(df2)
```

```python
df1.join(df2, rsuffix='C')
```

```python
df1.join(df2, lsuffix='C')
```

```python
jdf1 = df1.join(df2, lsuffix='sab')
jdf1
```

```python
jdf2 = jdf1.join(df3)
jdf2
```

```python
from datetime import datetime

ts = data['timeStamp'].astype(str)
new_ts=[]
for i in ts:
    i2=i.split()[0]
    nts = datetime.strptime(i2, '%Y-%m-%d')
    new_ts.append(nts.strftime('%a %b %d %Y'))
new_ts
```

```python
data['New_timeStamp'] = pd.DataFrame(new_ts)
data
```
